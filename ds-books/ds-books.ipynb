## Top Book

```python
# look fo book with search world, i.e. 'python'
def look_for_book(word):
    py_book = df1[df1['title'].str.contains(word)]
    top_py_book = py_book.nlargest(5, ['n_reviews'])
    return top_py_book
```

## Clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# verctorizing text
vec = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
X = vec.fit_transform(df1['title'])
```

```python
# vector to DataFrame
pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

### K-Means

---

K-Means puts out an centroid at an random location. Then iterates to find the center of a cluster. However, the user needs to manually set how many clusters we want to look at.

```python
from sklearn.cluster import KMeans
```

```python
# running k means
# with a minimum of 2 clusters
sum_sqrt_dist = []
K = range(2,20)
for k in K:
    k_m = KMeans(n_clusters=k, max_iter=600, n_init=10)
    k_m.fit(X) # X is the vectorized text in pd format
    sum_sqrt_dist.append(k_m.inertia_)
```

```python
# plotting elbow method for the cluster
plt.plot(K, sum_sqrt_dist, 'bx-')
plt.xlabel('Num of Cluster')
plt.ylabel('Sum of sqrt dist')
plt.title('Elbow Method: optimal K')
plt.show()
```
