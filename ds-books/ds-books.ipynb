## Clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# verctorizing text
vec = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
X = vec.fit_transform(df1['title'])
```

```python
# vector to DataFrame
pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

### K-Means

---

K-Means puts out an centroid at an random location. Then iterates to find the center of a cluster. However, the user needs to manually set how many clusters we want to look at.

```python
from sklearn.cluster import KMeans
```

```python
# running k means
# with a minimum of 2 clusters
sum_sqrt_dist = []
K = range(2,10)
for k in K:
    k_m = KMeans(n_clusters=k, max_iter=600, n_init=10)
    k_m.fit(X) # X is the vectorized text in pd format
    sum_sqrt_dist.append(k_m.inertia_)
```

```python
# plotting elbow method for the cluster
plt.plot(K, sum_sqrt_dist, 'bx-')
plt.xlabel('Num of Cluster')
plt.ylabel('Sum of sqrt dist')
plt.title('Elbow Method: optimal K')
plt.show()
```

The elbow method helps us to determin the optimal number of clusters. The most optimal cluster can be called 'True Cluster' to make it easier to diffriente.

By observing the graph above, we can see that there are some "bending" ad 8 cluster. Hence therefore we go with 8 as True Cluster.

```python
# True Cluster
t_k = 8
model = KMeans(n_clusters=t_k, init='k-means++', max_iter=600, n_init=10)
model.fit(X)

# pred/labels
labels = model.labels_
cluster_book = pd.DataFrame(list(zip(df1['title'], labels)), columns=['title', 'cluster'])
cluster_book_sort = cluster_book.sort_values(by=['cluster'])
print(cluster_book_sort)
```

Visualization of the clusters with the library wordcloud.

```python
from wordcloud import WordCloud
```

```python
for k in range(t_k):
    text = cluster_book[cluster_book.cluster == k]['title'].str.cat(sep=' ')
    wc = WordCloud(max_font_size=50, max_words=100, background_color='white').generate(text)
    # plotting 
    plt.subplot(2,4,k+1).set_title('Cluster ' + str(k))
    plt.plot()
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
plt.show()
```

```python
def get_review_url(product_url):
    try:
        split_url = product_url.split('dp')
        product_number = split_url[1].split('/')[1]
        review_url = split_url[0] + 'product-reviews/' + product_number + "/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
    except:
        review_url = None
    return review_url
```

```python
# generating urls for each revew
df1['review_urls'] = df1['complete_link'].apply(lambda x: get_review_url(x))
# cleaning the data by removing null-values
df_reviews = df1.loc[~df1['review_urls'].isnull()].reset_index()
df_reviews.head()
```
