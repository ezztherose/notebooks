Following the exemples of: https://pandas.pydata.org/docs/user_guide/10min.html


```python
import pandas as pd
import numpy as np
```

```python
# Create a series of values 
# Panda creats an index
s = pd.Series([1,2,3, np.nan, 6, 8])
s
```

```python
# Create a dataframe by passing in a NumPy array
# DataFrame contains: datetime, index & label for colums
dates = pd.date_range("20210815", periods=6)
dates

df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list("ABCD"))
df
```

```python
# Create DataFrame by passing
# dict (dictionary) of obj (objects) 
# that can convert to a series-like

"""
problem with this, 
check the spaceing
"""
df2 = pd.DataFrame(
     {
         "A": 1.0,
         "B": pd.Timestamp("20130102"),
         "C": pd.Series(1, index=list(range(4)), dtype="float32"),
         "D": np.array([3] * 4, dtype="int32"),
         "E": pd.Categorical(["test", "train", "test", "train"]),
         "F": "foo",
     } )
df2
```

```python
df2.dtypes
```

```python
# using IPython, tab completion for column names (as well as public attributes) is automatically enabled
df2.<TAB> # Don't work in iPython
```

```python
"""
DataFrame.to_num,py()
---------------------
* Gives NumPy representation of underlying data
* NOTE! can be an expensive operation if DataFrame has columns have diffrent data types
* NumPy: NumPy arrays have one dtype for the entire array, while pandas DataFrames have one dtype per column
* DataFrame.to_numpy(): 
    pandas will find the NumPy dtype that can hold all of the dtypes in the DataFrame. This may end up being object,
    which requires casting every value to a Python object
"""

# 'df', DataFram of floating-point values, DataFram.to_numpy() is fast and don't require copying data
df.to_numpy()
```

```python
# 'df2', DataFrame with multiple dtypes. DataFram.to_noimpy() is relatively expensive
df2.to_numpy()
```

Selection
---

- Getting
- Selection by label
- Selection by position
- Boolean index
- Setting

```python
# ---------------
# *** GETTING ***
# ---------------

# selecting a single column
# gives a 'Series' of 
df["A"]
```

```python
# selecting via [], slices the row
df[0:3]
```

```python
# --------------------------
# *** SELECTION BY LABEL ***
# --------------------------

# getting cross section by using an label
df.loc[dates[0]]
```

```python
# selection by multi-axis
df.loc[:,["A", "B"]]
```

```python
# Showing label slicing, both endpoints are included
df.loc["20210815":"20210820", ["A", "B"]]
```

```python
# Reduction in the dimensions of the returned object
df.loc["20210815", ["A", "B"]]
```

```python
# getting a scalar value
df.loc[dates[0], "A"]
```

```python
# getting fast access to a scalar (equivalent to the prior method)
df.at[dates[0], "A"]
```

```python
# -----------------------------
# *** Selection by position ***
# -----------------------------

# select via posaition of passed ints
df.iloc[3]
```
