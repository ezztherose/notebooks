```python
"""
Simple python
basic to reference to
"""
a = 5
#printing numbers
print(f'a is {a}') # pref
print('a is {}'.format(a))
print('a is ' + str(a))
print('a is %d' % (a))
```

```python
"""
if/else statements
"""

a = 5
b = 6

# 1
if a>5: 
    print('Value is grater than 5')
else:
    print('Value is not greater than 5')

# 2
if a ==5:
    print('Value is 5')
    if b==6:
        print('Value is also 6')
elif a==6:
    print('The value is 6')
```

```python
"""
Loops
"""
for x in range(1, 10):
    print(x)
    
acc = 0
for x in range(1, 10):
    acc += x
    print(f"adding {x}, sum {acc}")
    print(f"total: {acc}")
```

###  Python Lists, Dictionaries, Sets and JSON
- **Dictionary** - A dictionary is a mutable unordered collection that Python indexes with name and value pairs.
- **List** - A list is a mutable ordered collection that allows duplicate elements.
- **Set** - A set is a mutable unordered collection with no duplicate elements.
- **Tuple** - A tuple is an immutable ordered collection that allows duplicate elements.
