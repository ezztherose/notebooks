###  Python Lists, Dictionaries, Sets and JSON
- **Dictionary** - A dictionary is a mutable unordered collection that Python indexes with name and value pairs.
- **List** - A list is a mutable ordered collection that allows duplicate elements.
- **Set** - A set is a mutable unordered collection with no duplicate elements.
- **Tuple** - A tuple is an immutable ordered collection that allows duplicate elements.

---

- **append** - works with lits and touples
- **add** - is used when working with set

````python
"""
List and Touple
"""
l = ['a','b','c','d']
t = ('a','b','c','d')
print(l,t)

l[1] = 'changed'
print(l)

for s in l:
    print(s)

for i,l in enumerate(l):
    print(f"{i}:{l}")
````

````python
# Manually addint items to list
c = []
c.append('a')
c.append('b')
c.append('c')
c.append('d')
print(c)
print(c[1])
````

````python
# insert
c = ['a','b','c']
c.insert(0, 'a0')
print('insert ', c)

# remove
c.remove('b')
print('remove', c)

#remove at index
del c[0]
print('remove at index', c) 
````

````python
# Sets
s = set()
s = {'a','b','c'}
s = set(['a','b','c'])
print(s)

"""
- don't allow duplicates
- unordered collection
"""
````

````python
# Maps / Dictionaries / Hash Tables
d = {'name': "Ezz",'address': "444 Main"}
print(d)
print(d['name'])

if 'name' in d:
    print("Name defined")

if 'age' in d:
    print("Age defined")
else:
    print("Age undefined")

d.get('unknown_key', 'default')

# Access with individual keys and values of a dictionary'
print(f"Key: {d.keys()}") # all keys
print(f"Values: {d.values()}") # all values
````

**zip** --> combinding two list into one

````python
# python list & map structures
customers = [
    {"name": "Ezz & Roze", "pets": ["dumby", "puup", "jacky"]},
    {"name": "Tomato", "pets": ["jocky"]},
    {"name": "jane doe"}
]
print(customers)
print("-----------")
for customer in customers:
    print(f"{customer['name']}:{customer.get('pets', 'no pets')}")
````

````python
# Advanced lists
a = [1,2,3,4,5]
b = [5,4,3,2,1]

print(zip(a,b))
print(list(zip(a,b)))

for x,y in zip(a,b):
    print(f"{x} - {y}")

c = [5,4,3]
print(list(zip(a,c)))
````

````python
# enumerate to track index
a = [2,10,3,11,10,3,2,1]
for i, x in enumerate(a):
    if x>5:
        a[i] = 5
print(a)

# myltiply each iteration with 10, from 0 - 9
lst = [x*10 for x in range(10)]
print(lst)

"""
dict_variable = {key:value for (key,value) in dictonary.items()}
"""

text = ['col-zero', 'col-one', 'col-two', 'col-three']
lookup = {key:value for (value,key) in enumerate(text)}
print(lookup)
print(f"index of col-two {lookup['col-two']}")
````
